## Final

```python
risk_features = [
#positives
    "vlm_population_density_score",
    "yolo_total_vehicles",
    "vlm_footpath_score",
    "vlm_lane_marking_quality",
    "yolo_total_pedestrians",
    "vlm_road_condition_score",
    "yolo_total_bicycles",

# negatives
    "vlm_lane_count",
    # "vlm_curvature_visible",
    "vlm_intersection_visible",
    "yolo_total_motorcycles"
    # "vlm_overall_walkability"
    # "vlm_building_density_score",
]

df_risk_model = df_variable_maharashtra[risk_features]
print(df_risk_model.shape)
df_risk_model.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

scaler = MinMaxScaler()

df_norm = pd.DataFrame(
    scaler.fit_transform(df_risk_model[risk_features]),
    columns=risk_features,
    index=df_risk_model.index
)

df_norm.head()
```

```python
# Reverse lane count because more lanes generally reduce risk
df_norm["vlm_lane_count"] = 1 - df_norm["vlm_lane_count"]
df_norm["vlm_road_condition_score"] = 1 - df_norm["vlm_road_condition_score"]
df_norm["vlm_lane_marking_quality"] = 1 - df_norm["vlm_lane_marking_quality"]
```

```python
# Using the same weights as Thailand

weights = {
    "vlm_population_density_score": 0.24,   # coef = +0.5799 (strongest positive)
    "yolo_total_vehicles": 0.20,            # coef = +0.3520 (strong positive)
    "vlm_footpath_score": 0.08,             # coef = +0.1557 (moderate positive, but not that important as others)

    "vlm_lane_marking_quality": 0.04,       # coef = +0.0663 (weak positive)
    "yolo_total_pedestrians": 0.11,         # coef = +0.0523 (weak in regression, but high domain importance)
    "vlm_road_condition_score": 0.04,       # coef = +0.0307 (weak positive)
    "yolo_total_bicycles": 0.04,            # coef = +0.0218 (weak positive)

    "vlm_lane_count": 0.04,                 # coef = -0.0327 (important in real life context))
    # "vlm_curvature_visible": excluded      # coef = -0.0335 (too weak / noisy)
    "vlm_intersection_visible": 0.08,       # coef = -0.0686 (kept due to engineering logic)
    "yolo_total_motorcycles": 0.13,         # coef = -0.0937 (overrode regression; important in Thailand)

    # "vlm_overall_walkability": excluded    # coef = -0.3150 
    # "vlm_building_density_score": excluded # coef = -0.5126 (noisy)
}

df_risk_model["image_risk_score"] = sum(
    df_norm[col] * weight
    for col, weight in weights.items()
)
```
